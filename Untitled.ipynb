```python
import random
import time

start_time = time.time()

def rootmax(tab, n, i):
    plus_grand = i
    gauche = 2 * i + 1
    droite = 2 * i + 2

    if gauche < n and tab[gauche] > tab[plus_grand]:
        plus_grand = gauche
    if droite < n and tab[droite] > tab[plus_grand]:
        plus_grand = droite

    if plus_grand != i:
        tab[i], tab[plus_grand] = tab[plus_grand], tab[i]
        rootmax(tab, n, plus_grand)

def tas_sort(tab):
    n = len(tab)

   
    for i in range(n//2 - 1, -1, -1):
        rootmax(tab, n, i)

    
    for i in range(n - 1, 0, -1):
        tab[i], tab[0] = tab[0], tab[i]  
        rootmax(tab, i, 0)






end_time = time.time()
execute_time =end_time - start_time
with open("TAS_reel.txt","a") as f:
    # f.write("pour 10000 eleme")
    f.write(f"le temp execution :{execute_time}")

    f.write("\n")
print(start_time)
print(end_time)

print(execute_time)
listt=[random.randint(1,10000) for i in range(10000)]
print(listt)
tas_sort(listt)
print(listt)
```

```python
import random
import time

start_time = time.time()

def tri_selection(liste):
    n = len(liste)
    for i in range(n):
        min_index = i
        

        for j in range(i + 1, n):
            if liste[j] < liste[min_index]:
                min_index = j
        
  
        liste[i], liste[min_index] = liste[min_index], liste[i]
    
    return liste




end_time = time.time()
execute_time =end_time - start_time
with open("TAS_reel.txt","a") as f:
    # f.write("pour 10000 eleme")
    f.write(f"le temp execution :{execute_time}")

    f.write("\n")
print(start_time)
print(end_time)

print(execute_time)
listt=[random.randint(1,100) for i in range(100)]
print(listt)
tri_selection(listt)
print(listt)
```

```python
import random
import time

start_time = time.time()

def tri_insertion(tableau):
    for i in range(1, len(tableau)):        
        cle = tableau[i]                    
        j = i - 1    # ici le i :commence par 1 alor que le j :commence pa c aucun contraduction                      

     
        while j >= 0 and tableau[j] > cle:
            tableau[j + 1] = tableau[j]     
            j = j - 1                       

       
        tableau[j + 1] = cle




end_time = time.time()
execute_time =end_time - start_time
with open("time_100_a_bull.txt","a") as f:
    f.write(f"le temp execution :{execute_time}")
    f.write("\n")
# print(start_time)
# print(end_time)

# print(execute_time)
listt=[random.randint(1,1000) for i in range(1000)]
print(listt)
tri_insertion(listt)
print(listt)
```
